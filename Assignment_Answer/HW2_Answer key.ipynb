```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
turkey_pop = [20,30,40,50,60,70,80]
```

```python
plt.plot(years, iran_pop,ls='-',marker='+',mew=8)
plt.plot(years,turkey_pop,ls='--')
plt.title('Iran vs Turkey Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([20,30,40,50,60,70,80],['20M', '30M', '40M', '50M' ,'60M', '70M', '80M',])
plt.legend(['iran','Turkey'],loc='best')
plt.grid()
plt.show()
```
